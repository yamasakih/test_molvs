```python
from molvs import Standardizer
from rdkit import Chem
from rdkit.Chem import Draw
```

```python
smiles_set = [
    '[Na]OC(=O)c1ccc(C[S+2]([O-])([O-]))cc1',
    'O=C([O-])c1ccc(C[S](=O)=O)cc1.[Na+]',
    'O=C([O-])c1ccc(C[S](=O)=O)cc1',
]
mols = [Chem.MolFromSmiles(smiles) for smiles in smiles_set]
```

```python
Draw.MolsToGridImage(mols)
```

```python
standardizer = Standardizer()
standardized_mols = [standardizer.charge_parent(mol) for mol in mols]
```

```python
Draw.MolsToGridImage(standardized_mols)
```
